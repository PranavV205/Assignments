# Q1. What is multiprocessing in python? Why is it useful?

Multiprocessing refers to the ability of a system to support more than one processor at the same time. Applications in a multiprocessing system are broken to smaller routines that run independently. The operating system allocates these threads to the processors improving performance of the system.

Multiprocessing is useful in Python for several reasons:

1. Parallelism: Multiprocessing enables parallel execution of tasks, taking advantage of multiple CPU cores. This can lead to significant performance improvements, especially for CPU-bound tasks.

2. Isolation: Each process has its own memory space, which provides better isolation between tasks. This reduces the risk of data corruption and makes it easier to manage shared resources.

3. Avoiding Global Interpreter Lock (GIL): In Python, the Global Interpreter Lock (GIL) prevents multiple native threads from executing Python bytecodes at once, limiting the parallelism achievable through multithreading. Multiprocessing allows bypassing the GIL, as each process has its own interpreter and memory space.

4. Fault Tolerance: If one process crashes, it does not necessarily affect other processes, enhancing the overall stability of the program.

5. Scalability: Multiprocessing can be used to scale up computations on multiple machines, making it suitable for distributed computing.

# Q2. What are the differences between multiprocessing and multithreading?

1. Multiprocessing uses two or more CPUs to increase computing power, whereas multithreading uses a single process with multiple code segments to increase computing power.

2. Multithreading focuses on generating computing threads from a single process, whereas multiprocessing increases computing power by adding CPUs.

3. Multiprocessing is used to create a more reliable system, whereas multithreading is used to create threads that run parallel to each other.

4. Multithreading is quick to create and requires few resources, whereas multiprocessing requires a significant amount of time and specific resources to create.

5. Multiprocessing executes many processes simultaneously, whereas multithreading executes many threads simultaneously.

6. Multithreading uses a common address space for all the threads, whereas multiprocessing creates a separate address space for each process.

# Q3. Write a python code to create a process using the multiprocessing module.

```python
import multiprocessing
import time

start = time.perf_counter()

def do_something():
  print('Sleeping 1 second...')
  time.sleep(1)
  print('Done sleeping...')

processes = []

for _ in range(2):
  p = multiprocessing.Process(target = do_something)
  p.start()
  processes.append(p)

for process in processes:
  process.join()

finish = time.perf_counter()

print(f'Finished in {round(finish - start, 2)} second(s)')
```

# Q4. What is a multiprocessing pool in python? Why is it used?

A process pool is a programming pattern for automatically managing a pool of worker processes.

The pool is responsible for a fixed number of processes.

It controls when they are created, such as when they are needed.
It also controls what they should do when they are not being used, such as making them wait without consuming computational resources.

Python provides a process pool via the multiprocessing.Pool class.

There are four main steps in the life-cycle of using the multiprocessing.Pool class, they are: create, submit, wait, and shutdown.

1. Create: Create the process pool by calling the constructor multiprocessing.Pool().
2. Submit: Submit tasks synchronously or asynchronously.
3. Wait: Wait and get results as tasks complete (optional).
4. Shutdown: Shutdown the process pool by calling shutdown().

Multiprocessing pools are used when:

1. Tasks can be defined by a pure function that has no state or side effects.

2. Your task can fit within a single Python function, likely making it simple and easy to understand.

3. You need to perform the same task many times, e.g. homogeneous tasks.

4. You need to apply the same function to each object in a collection in a for-loop.

# Q5. How can we create a pool of worker processes in python using the multiprocessing module?


In Python, you can create a pool of worker processes using the multiprocessing module's Pool class. The Pool class provides a convenient way to parallelize the execution of a function across multiple input values, distributing the work among a specified number of worker processes.

```python
import multiprocessing

def square_number(x):
    return x * x

if __name__ == "__main__":
    # Create a Pool with 3 worker processes
    with multiprocessing.Pool(processes=3) as pool:
        # Input values for the square_number function
        input_values = [1, 2, 3, 4, 5]

        # Use the map function to apply square_number to each input value in parallel
        results = pool.map(square_number, input_values)

    # Print the results
    print("Input Values:", input_values)
    print("Squared Values:", results)
```

Another way to create a pool of worker processes in python can be achived by importing the concurrent.futures module.

# Q6. Write a python program to create 4 processes, each process should print a different number using the import multiprocessing

```python
import multiprocessing

def print_number(number):
    print(f"Process ID: {multiprocessing.current_process().pid}, Number: {number}\n")

if __name__ == "__main__":
    # Create a list of numbers
    numbers = [1, 2, 3, 4]

    # Create a Pool with 4 worker processes
    with multiprocessing.Pool(processes=4) as pool:
        # Use the map function to apply print_number to each number in parallel
        pool.map(print_number, numbers)
```
